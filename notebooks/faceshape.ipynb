```python
import os

#set base dir to extracted dataset
base_dir = "C:\\Users\\Loq\\OneDrive\\Documents\\face\\DATASET"
#list all subfolders(my classes)
classes = [folder for folder in os.listdir(base_dir)]
print(classes)
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import random
```

```python
class_img = {}
class_length = {}

for cls in classes:
    cls_path = os.path.join(base_dir,cls)
    class_img[cls] = [sample for sample in os.listdir(cls_path)]
    class_length[cls] = len(class_img[cls])
    print(f"{cls} : {class_length[cls]}")
```

```python
print(class_img)
```

```python
label=classes[0]
sample_file=class_img[label][0]
print(label)
print(sample_file)
```

```python
img_path=os.path.join(base_dir,label,sample_file)
```

```python
img=Image.open(img_path)
plt.imshow(img)
plt.title(label)
plt.axis("off")
plt.show()
```

```python
ext=(".jpg",".jpeg",".png")
for cls in classes:
    files=[file for file in os.listdir(os.path.join(base_dir,cls)) if file.endswith(ext)]
    sample_file=files[0]
    img_path=os.path.join(base_dir,cls,sample_file)
    img=Image.open(img_path)
    plt.imshow(img)
    plt.title(cls)
    plt.axis("off")
    plt.show()
```

```python

plt.bar(class_length.keys(),class_length.values())
plt.xlabel("Classes")
plt.ylabel("Number of images")
plt.title("Distribution of images across classes")
plt.show()
```
